# Quick start
This notebook explains the format of the data and how to plot it. 

```python
import pandas as pd
from matplotlib import pyplot as plt
```

```python
# Select path to read data from
rt = '../data/data.csv'
```

```python
# Load and display data
# Note that 'time' is in seconds. Every row in the dataframe represents a new sample, with one sample per minute.
# Data in each column as already been normalized.
# There are 48 functions in the dataset, each with data for requests, cpu, pods, cost (execution time) and memory
# This is a total of 48 functions x 5 metrics per function = 240 time series.
df = pd.read_csv(rt)
df
```

```python
# Print length of dataset
print(f'This dataset is {len(df)/1440} days long')
```

```python
# Data does not contain NaNs
df.isna().sum().sum()
```

```python
# Plot a time series

target = '13_requests'

fig, ax = plt.subplots(1,1,figsize=(12,3))
plt.scatter(df['time'], df[target], s=1)
plt.title(target, fontsize=20)
plt.xlabel('Time (seconds)')
plt.show()
```
